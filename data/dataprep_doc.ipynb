# Class Imbalance
Find out how many objects per class there are in total

```python
import dataprep_funcs

root="/teamspace/studios/this_studio/data/dota-mod"
mydota=dataprep_funcs.DotaDataset_1p5(root=root, folder="train")
class_count = mydota.instances_per_class()
files_per_class = mydota.files_per_class()
```

```python
dota_classes={
    1: "plane", 
    2: "ship", 
    3: "storage tank", 
    4: "baseball diamond", 
    5: "tennis court", 
    6: "basketball court", 
    7: "ground track field", 
    8: "harbor", 
    9: "bridge", 
    10: "large vehicle", 
    11: "small vehicle", 
    12: "helicopter", 
    13: "roundabout", 
    14: "soccer ball field", 
    15: "swimming pool",
    16: "container crane"}
```

```python
import matplotlib.pyplot as plt
print(class_count)
print(len(class_count.keys()))

# plot bar graph
categories = [dota_classes[x + 1] for x in class_count.keys()]
vals = [x for x in class_count.values()]
plt.bar(categories, vals, color="skyblue")
plt.yscale('log')
plt.ylabel('Count')
plt.xlabel('Categories')
plt.xticks(rotation=90)
plt.title('DOTA 1.5 Class Count')
plt.show()
```

```python
print(files_per_class)
```

```python
# plot bar graph for number of files per class
categories = [dota_classes[x + 1] for x in files_per_class.keys()]
vals = [x for x in files_per_class.values()]
plt.bar(categories, vals, color="skyblue")
plt.yscale('log')
plt.ylabel('# of Images')
plt.xlabel('Categories')
plt.xticks(rotation=90)
plt.title('DOTA 1.5 # of Images Associated Per Category')
plt.show()
```
